```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\PGA-43\Power BI\PowerBI Project made\Burgetto's_orders1.csv")
```

```python
data.head()
```

```python
# 1) Total Revenue
total_revenue = data['Total_price'].sum()
print(f"Total Revenue: {total_revenue}")
```

```python
# 2) Top 5 Items by Revenue
data.groupby('item_name')['Total_price'].sum().nlargest(5).plot(kind='bar', title='Top 5 Items by Revenue')
plt.show()
```

```python
# 3) Top 5 Items by Quantity Sold
data.groupby('item_name')['quantity'].sum().nlargest(5).plot(kind='bar', title='Top 5 Items by Quantity')
plt.show()
```

```python
# 4) Sales by Item Category
data.groupby('item_category')['Total_price'].sum().plot(kind='bar', title='Sales by Item Category')
plt.show()
```

```python
# 5) Sales by Item Size
data.groupby('item_size')['Total_price'].sum().plot(kind='bar', title='Sales by Item Size')
plt.show()
```

```python
# 6) Average Order Value
print("Average Order Value:", data['Total_price'].sum() / data['order_id'].nunique())
```

```python
# 7) Total Number of Unique Orders
print("Total Orders:", data['order_id'].nunique())
```

```python
# 8) Most Frequent Item Category
print("Most Frequent Category:", data['item_category'].value_counts().idxmax())
```

```python
# 9) Least Popular Items by Quantity
data.groupby('item_name')['quantity'].sum().nsmallest(5).plot(kind='bar', title='Least Popular Items by Quantity')
plt.show()
```

```python
# 10) Top 5 Items by Revenue
top_5_revenue = data.groupby('item_name')['Total_price'].sum().nlargest(5)
print("Top 5 Items by Revenue:")
print(top_5_revenue)
```

```python
# 11) Top 5 Items by Quantity
top_5_quantity = data.groupby('item_name')['quantity'].sum().nlargest(5)
print("Top 5 Items by Quantity:")
print(top_5_quantity)
```

```python
# 12) Top 5 Items by Total Orders
top_5_orders = data.groupby('item_name')['order_id'].nunique().nlargest(5)
print("Top 5 Items by Total Orders:")
print(top_5_orders)
```

```python
# 13) % of Sales by Item Category
sales_by_category = data.groupby('item_category')['Total_price'].sum()
sales_by_category_percentage = (sales_by_category / total_revenue * 100).round(2)
print("% of Sales by Item Category:")
print(sales_by_category_percentage)
```

```python
# 14) % of Sales by Item Size
sales_by_size = data.groupby('item_size')['Total_price'].sum()
sales_by_size_percentage = (sales_by_size / total_revenue * 100).round(2)
print("% of Sales by Item Size:")
print(sales_by_size_percentage)
```

```python
# 15) Average Items Per Order
avg_items_per_order = round(data['quantity'].sum() / data['order_id'].nunique(), 2)
print(f"Average Items Per Order: {avg_items_per_order}")
```
