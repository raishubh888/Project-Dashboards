```python
# Now creating graph minimum to maximum sales w.r.t hire date.

import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(dt)
```

```python
# Convert hire_date to datetime
df['hire_date'] = pd.to_datetime(df['hire_date'])

# Group by hire_date and sum sales
sales_by_date = df.groupby('hire_date')['sales'].sum()

# Sort by hire_date in ascending order
sales_by_date = sales_by_date.sort_index()
```

```python
# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(sales_by_date.index, sales_by_date.values)
plt.xlabel('hire_date')
plt.ylabel('sales in million usd')
plt.title('Sales w.r.t Hire Date')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#Q2) Which job role has maximum sales with respect to marital status?

#Soln2) 

df = pd.DataFrame(dt)
```

```python
# Group by job role and marital status, and calculate total sales
grouped = df.groupby(["job_role", "maritalstatus"])["sales"].sum()

# Find the job role with maximum sales
max_sales_job_role = grouped.idxmax()

print(f"The job role with maximum sales is: {max_sales_job_role[0]} (Marital Status: {max_sales_job_role[1]})")
```

```python
#Q3) Which job role has maximum yearly income ascending to descending ?

# Sort the DataFrame by 'yearlyincome' in descending order
sorted_df = df.sort_values(by='yearlyincome', ascending=False)

# Display the sorted DataFrame
print(sorted_df[['job_role', 'yearlyincome']])
```

```python
# Sort the DataFrame by 'yearlyincome' in descending order
sorted_df = df.sort_values(by='yearlyincome', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_df['job_role'], sorted_df['yearlyincome'])
plt.xlabel('Job Role')
plt.ylabel('Yearly Income in thousand usd')
plt.title('Maximum Yearly Income by Job Role')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Q4) What education one should persue to have maximum yearly income ?

#sol) # Group by 'education' and calculate mean yearly income
average_income_by_education = dt.groupby('education')['yearlyincome'].mean()

# Find the education with the maximum average yearly income
max_income_education = average_income_by_education.idxmax()

print(f"The education one should pursue for maximum yearly income is: {max_income_education}")
```

```python
#Q5) Which education persued has minimum sales?

#sol) # Group by 'education' and calculate minimum sales
min_sales_by_education = dt.groupby('education')['sales'].min()

# Find the education with the minimum sales
min_sales_education = min_sales_by_education.idxmin()

print(f"The education pursued with minimum sales is: {min_sales_education}")
```

```python
# Group by 'education' and calculate minimum sales
min_sales_by_education = dt.groupby('education')['sales'].min()

# Sort the data by minimum sales in ascending order
sorted_min_sales = min_sales_by_education.sort_values()

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_min_sales.index, sorted_min_sales.values)
plt.xlabel('Education')
plt.ylabel('Sales in thousand usd')
plt.title('Minimum Sales by Education')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Q6) What job role has maximum sales? why do you think it did?

#Soln) # Group by 'job_role' and calculate maximum sales
max_sales_by_job_role = dt.groupby('job_role')['sales'].max()

# Find the job role with the maximum sales
max_sales_job_role = max_sales_by_job_role.idxmax()

print(f"The job role with maximum sales is: {max_sales_job_role}")
```
